DATA MANIPULATION IN PANDAS

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

DATA SET IMPORTED

```python
df = pd.read_csv(r"/content/student_data.csv")
df.head()
```

```python
df.info()
```

```python
# Re-name the column label in pandas :
df.rename(columns={'Date of Birth':'DOB','Student ID':"ID"},inplace=True)
```

```python
# Remove the empty cell :
df = df.dropna()
```

***Student first name and second name are placed in new column.***

```python
df[['First Name', 'Last Name']] = df['Student Name'].str.split(' ', n=1, expand=True)
df.head()
```

***student name are sorted in Alphabetical order***




```python
df["Student Name"] = sorted(df["Student Name"])
df.head()
```

***created a new column that can return the current age of a student***

```python
df['DOB'] = pd.to_datetime(df['DOB']) # In Date of Birth column

current_date = pd.to_datetime(datetime.now().strftime('%Y-%m-%d')) # using datetime.now()

df['Age'] = (current_date - df['DOB']).astype('<m8[Y]')
df.head()
```

***Get the age group with the highest count***

```python
age_counts = df['Age'].value_counts() # list of age and their count

most_common_age = age_counts.idxmax() # here shows among the age_count which age has higher count

most_common_age_count = df[df["Age"] == 21] # this line prints the most age common count
most_common_age_count = most_common_age_count.value_counts().sum()


print(most_common_age,"Aged peoples are larger in count : ",most_common_age_count)
```

***shown No of student on their feeld of study***

```python
df.groupby("Field of Study")["ID"].count()
```

***Created a pevert table that shows Average Fees by Field of Study, Year of Admission, and Specialization***

```python
pivot_table = pd.pivot_table(df, values='Fees', index=['Field of Study','Year of Admission'], columns='Specialization', aggfunc='mean')
pivot_table
```
